````python
a = list(map(int, input().split()))
d = 0
c = 0
q = 10**99
e = 0

for i in a:
    if d < i:
        d = i
        c += 1
    else:
        break

for i in a:
    if q > i:
        q = i
        e += 1
    else:
        break

if len(a) == c:
    print("tang")
elif len(a) == e:
    print("giam")
else:
    print("khong")
````

````python
def find_max(a):
    max = a[0]
    for i in a:
        if i > max:
            max = i
    return max

a = list(map(int, input().split()))
print(find_max(a))
````

````python
def find_avg(a):
    sum = 0
    for i in a:
        sum += i
    return sum / len(a)

a = list(map(int, input().split()))
print(find_avg(a))
````

````python
def is_palindrome(a):
    for i in range(len(a) // 2):
        if a[i] != a[len(a)-1-i]:
            return False
    return True

a = list(map(int, input().split()))
if is_palindrome(a):
    print("yes")
else:
    print("no")
````

````python
def move_zeros(a):
    result = []
    zeros = []
    for i in a:
        if i == 0:
            zeros.append(i)
        else:
            result.append(i)
    return zeros + result

a = list(map(int, input().split()))
print(move_zeros(a))
````
