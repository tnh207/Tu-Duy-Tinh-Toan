```python
def tinh_phan_tu_o_vi_tri_chan(day_so_nguyen):
    phan_tu_o_vi_tri_chan = [day_so_nguyen[i] for i in range(len(day_so_nguyen)) if i % 2 == 0]
    return phan_tu_o_vi_tri_chan

# Ví dụ sử dụng:
day_so_nguyen = [1, 2, 3, 4, 5, 6]
print("Phần tử ở vị trí chẵn trong dãy số nguyên:", tinh_phan_tu_o_vi_tri_chan(day_so_nguyen))
```
