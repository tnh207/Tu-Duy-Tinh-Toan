## W6A1 – Loại bỏ phần tử trùng (giữ nguyên thứ tự)

Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.

Yêu cầu in ra đúng dạng list Python, ví dụ `[1, 2, 3]`

```python
nums = list(map(int, input().split()))
seen = set()
result = []
for x in nums:
    if x not in seen:
        seen.add(x)
        result.append(x)
print(result)
```

## W6A2 – Tổng tích lũy (Cumulative Sum)

Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu tới vị trí i), giữ nguyên thứ tự.

```python
nums = list(map(int, input().split()))
cumsum = []
total = 0
for x in nums:
    total += x
    cumsum.append(total)
print(cumsum)
```

## W6A3 – Xoay trái tuple k vị trí

Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp xếp.)

```python
t = tuple(map(int, input().split()))
k = int(input())
k = k % len(t) if t else 0
result = t[k:] + t[:k]
print(result)
```

## W6A4 – Dictionary ánh xạ key → list các value

Nhập một dòng gồm các cặp `key:value` cách nhau bởi dấu cách. Tạo dictionary ánh xạ mỗi key → list các value theo thứ tự xuất hiện. Nếu key lặp lại, thêm vào list của key đó.

```python
pairs = input().split()
d = {}
for pair in pairs:
    key, value = pair.split(':')
    if key not in d:
        d[key] = []
    d[key].append(value)
print(d)
```

## W6A5 – Đếm số dương, âm, và bằng 0

Nhập một list số nguyên. In ra dictionary `{'positives': ..., 'negatives': ..., 'zeros': ...}` (đếm số >0, <0, =0).

```python
nums = list(map(int, input().split()))
result = {
    'positives': sum(1 for x in nums if x > 0),
    'negatives': sum(1 for x in nums if x < 0),
    'zeros':     sum(1 for x in nums if x == 0)
}
print(result)
```

## W6A9 – Gộp hai dictionary (cộng giá trị nếu trùng key)

Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại; nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới theo thứ tự từ điển tăng dần của key.

**Định dạng nhập:** Mỗi dòng là một dictionary dạng `key1:value1 key2:value2 ...`

```python
def parse_dict(line):
    d = {}
    for pair in line.split():
        k, v = pair.split(':')
        d[k] = int(v)
    return d

d1 = parse_dict(input())
d2 = parse_dict(input())

merged = dict(d1)
for k, v in d2.items():
    merged[k] = merged.get(k, 0) + v

result = dict(sorted(merged.items()))
print(result)
```

## W6A10 – Các cặp số có tổng bằng k

Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.

```python
nums = list(map(int, input().split()))
k = int(input())

pairs = []
for i in range(len(nums)):
    for j in range(i + 1, len(nums)):
        if nums[i] + nums[j] == k:
            pairs.append((nums[i], nums[j]))

pairs.sort(key=lambda x: x[0])
print(pairs)
```

## W6A11 – Tách số chẵn và lẻ từ tuple

Nhập vào một tuple chứa các số nguyên. Hãy tạo ra hai tuple mới: một chứa tất cả các số chẵn và một chứa tất cả các số lẻ từ tuple ban đầu, giữ nguyên thứ tự tương đối. In ra hai tuple này trên hai dòng riêng biệt.

```python
t = tuple(map(int, input().split()))
evens = tuple(x for x in t if x % 2 == 0)
odds  = tuple(x for x in t if x % 2 != 0)
print(evens)
print(odds)
```

## W6A12 – Tìm phần tử xuất hiện nhiều nhất

Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ nhất trong số đó.

```python
nums = list(map(int, input().split()))
count = {}
for x in nums:
    count[x] = count.get(x, 0) + 1
max_freq = max(count.values())
candidates = [k for k, v in count.items() if v == max_freq]
print(min(candidates))
```

## W6A13 – Đảo ngược key và value của dictionary

Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo một dictionary mới bằng cách đảo ngược key và value của dictionary ban đầu. In ra dictionary mới.

**Định dạng nhập:** `key1:value1 key2:value2 ...`

```python
d = {}
for pair in input().split():
    k, v = pair.split(':')
    d[k] = v
reversed_d = {v: k for k, v in d.items()}
print(reversed_d)
```

## W6A15 – Lọc dictionary theo giá trị

Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp key-value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary mới.

**Định dạng nhập:** Dòng 1: `key1:value1 key2:value2 ...` | Dòng 2: số nguyên k

```python
d = {}
for pair in input().split():
    k, v = pair.split(':')
    d[k] = int(v)
threshold = int(input())
result = {k: v for k, v in d.items() if v > threshold}
print(result)
```
