```python
# 1. Khởi tạo danh sách sinh viên ban đầu
students = {
    'S001': {'name': 'An', 'score': 8.5},
    'S002': {'name': 'Binh', 'score': 9.0}
}

# 2. Thêm sinh viên mới vào dictionary
students['S003'] = {'name': 'Chi', 'score': 7.5}

# 3. In danh sách sinh viên ra màn hình
print("--- Student List ---")
for sid, info in students.items():
    print(f"ID: {sid} | Name: {info['name']} | Score: {info['score']}")

# 4. Tính điểm trung bình của tất cả sinh viên
if students: # Kiểm tra xem dictionary có dữ liệu không để tránh lỗi chia cho 0
    scores = [info['score'] for info in students.values()]
    average = sum(scores) / len(scores)
    
    print("-" * 20)
    print(f"Average score of all students: {average:.2f}")
else:
    print("No students in the directory.")
```
